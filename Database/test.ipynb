```python
from qdrant_client import QdrantClient
import numpy as np
from qdrant_client.models import PointStruct
import json
import torch
import clip
from PIL import Image

client = QdrantClient(host="localhost", port=6333)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)
```

```python
text_description = "Men's Long-Sleeve Button-Down Shirt (Brown). Priced at $34.99,"
```

```python
tokenized_text = clip.tokenize([text_description]).to(device)
```

```python
with torch.no_grad():
    text_embedding = model.encode_text(tokenized_text).cpu().numpy().flatten()
    print(f"Embedding vectors: {text_embedding.tolist()}")
```

```python
query_vector = text_embedding.tolist()
hits = client.search(
   collection_name="products_text",
   query_vector=query_vector,
   limit=5  # Return 5 closest points
)
```
